## Q3: Prove your previous result using mean, median and mode.

````python
#calculate measurements for every dataframe feature

measurements={"feature":[], "mean":[], "median":[], "mode":[]}

for i in df_boston:
    measurements["feature"].append(i)
    measurements["mean"].append(df_boston[i].mean())
    measurements["median"].append(df_boston[i].median())
    measurements["mode"].append(df_boston[i].mode().values[0])

prove_measurements =pd.DataFrame(measurements)
prove_measurements
````

## Q8: Simulate Two Coin Flips, then draw the result distribution.

````python
#define a function that will randomly return a choice, either heads or tails
def flips():
    return np.random.choice(['Heads', 'Tails'])

total_heads=0
total_tails=0

for i in range(5000):
    if flips()=='Heads':
        total_heads+=1
    else:
        total_tails+=1

sns.barplot(x=['Heads', 'Tails'], y=[total_heads,total_tails])
plt.xlabel('Coin Side')
plt.ylabel('Frequency')
plt.title('Coin Flips Distribution')
plt.show()
````

I used the same [Coin Flip Simulator](https://www.youtube.com/watch?v=FqfRhnbSdf8) code that was explained on youtube.
##### We can see that the two sides of the coin have 0.5 probability. Every time we rerun the code, the result of the two sides will be close.

# Q9: Summarize the differences between the Z-test and the T-test in three sentences.

1. The Z-test is used when the population standard deviation is known, while the T-test is used when the population standard deviation is unknown. 


2. The T-test is more powerful when the sample size is small. While the Z-test need the sample size to be large. 


3. The T-test has more flexibility in terms of choosing between one-sample, two-sample, and paired tests, while the Z-test is typically only used for one-sample tests.

4. The Z-test is more reliable when working with non-normally distributed data. Also, it has statistical power with large sample sizes compared to T-test.

## Calculate

Let's assume that a class's average score is higher than 70 with a standard deviation of 10.
Calculate the Z-value to determine whether there is sufficient data to support this claim at a 0.05 
significance level if a sample of 50 students was chosen and their mean score was 80.

````python
population_mean, standard_deviation, significance_level, sample_size, sample_mean = 70, 10, 0.05, 50, 80

Z_value = (sample_mean - population_mean) / (standard_deviation / (math.sqrt(sample_size))) #calculate Z-value

critical_value = stats.norm.ppf(1 - significance_level) #calculate the critical value

print("The Z-value is "+ str(Z_value) +" and the critical value is "+ str(critical_value))
````

#### Since the Z-value is greater than the critical value, we can say that there is sufficient data to support this claim at a 0.05 significance level.

An establishment wants to boost sales. According to past sales data, each sale made by 30 salesmen averaged 
40 dollar. The most recent data indicated that an average sale per transaction was 60 dollar after some training.Find the t-value for a 20 dollar standard deviation. Did a training program increase sales?

````python
sample_size, population_mean, sample_mean, standard_deviation = 30, 40, 60, 20

T_value = (sample_mean - population_mean) / (standard_deviation / (math.sqrt(sample_size))) #calculate T-value

critical_value = stats.t.ppf(1 - significance_level/2, sample_size-1) #calculate the critical value, ppf is Percent point function, and the parameter(critical level, degrees of freedom)

print("The T-value is "+ str(T_value) +" and the critical value is "+ str(critical_value))
````

#### Since the T-value is greater than the critical value, we can say that the training program increase sales.
